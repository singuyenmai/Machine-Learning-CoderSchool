```python
import pandas as pd
import numpy as np
```

```python
song_dataset = pd.read_json('../MasterSongList.json')
print('Raw data size: ', song_dataset.shape)
song_dataset.head()
```

```python
def consolidateGenres (genre):
    if (len(genre) > 0):
        gen = genre[0]
        return gen.split(':')[0]
    else:
        return np.nan
    
song_dataset.loc[:, 'genres']= song_dataset['genres'].apply(consolidateGenres)
```

```python
song_dataset.head()
```

```python
audioFeatures_names = ['key', 'energy', 'liveliness', 'tempo', 'speechiness', 'acousticness', 'instrumentalness', 
                  'time_signature', 'duration', 'loudness', 'valence', 'danceability', 'mode', 
                  'time_signature_confidence', 'tempo_confidence', 'key_confidence', 'mode_confidence']

def audioFeatureFrame (features_val, features_names):
    values = [val for val in features_val]
    frame = pd.DataFrame(values, columns = features_names)
    return frame
```

```python
audio_features_df = audioFeatureFrame(song_dataset.loc[:, 'audio_features'], audioFeatures_names)
audio_features_df
```

```python
from string import punctuation
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
from nltk.stem.snowball import SnowballStemmer

translator = str.maketrans('','', punctuation)
stemmer = SnowballStemmer('english')

def clean_text(raw_text):
    
    clean_words = []
    
    raw_text = raw_text.lower()
    raw_text = raw_text.translate(translator)
    
    for word in raw_text.split():
        
        if word not in ENGLISH_STOP_WORDS:
            
            clean_words.append(stemmer.stem(word))
    
    return " ".join(clean_words)
```

```python
lyrics_df = song_dataset.loc[:,'lyrics_features'].apply(lambda x: clean_text(" ".join(x)))
```
