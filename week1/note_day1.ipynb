```python
# for reading data
import pandas as pd
```

```python
raw = pd.read_json('MasterSongList.json')
print('Raw data size: ', raw.shape)
raw.head()
```

```python
raw['genres']
```

```python
def cleanGenres (genre):
    if (len(genre) > 0):
        gen = genre[0]
        return gen.split(':')[0]
    else:
        return "NA"
    
cleaned = raw['genres'].apply(cleanGenres)
cleaned
```
