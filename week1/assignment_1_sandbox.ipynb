```python
# for reading data
import pandas as pd
import numpy as np
```

```python
song_dataset = pd.read_json('MasterSongList.json')
print('Raw data size: ', song_dataset.shape)
song_dataset.head()
```

```python
song_dataset['genres']
```

```python
def consolidateGenres (genre):
    if (len(genre) > 0):
        gen = genre[0]
        return gen.split(':')[0]
    else:
        return "NA"
    
song_dataset.loc[:, 'genres']= song_dataset['genres'].apply(consolidateGenres)
song_dataset['genres']
```

```python
song_dataset.head()
```

```python
rock_pop_idx = (song_dataset['genres'] == 'rock') | (song_dataset['genres'] == 'pop')
rock_pop_dataset = song_dataset.loc[rock_pop_idx].copy()
rock_pop_dataset.reset_index(drop = True, inplace = True)
print(rock_pop_dataset.shape)
rock_pop_dataset.head()
```

```python
classes = ['pop', 'jazz', 'rock']
pop_jazz = song_dataset.query('genres in @classes')
pop_jazz
```

```python
def audioFeatureFrame (features_col, features_names):
    featureList = [features for features in features_col]
    frame = pd.DataFrame(featureList, columns = features_names)
    return frame
```

```python
audioFeatures_names = ['key', 'energy', 'liveliness', 'tempo', 'speechiness', 'acousticness', 'instrumentalness', 
                  'time_signature', 'duration', 'loudness', 'valence', 'danceability', 'mode', 
                  'time_signature_confidence', 'tempo_confidence', 'key_confidence', 'mode_confidence']
```

### Big dataset of rock and pop songs (6435 rock songs & 1889 pop songs), drop all NAs

```python
audioFeatures = audioFeatureFrame(rock_pop_dataset.loc[:, 'audio_features'], audioFeatures_names)
audioFeatures.head()
```

```python
audioFeatures_genres = pd.concat([audioFeatures, rock_pop_dataset.loc[:,'genres']], axis = 1)
```

```python
audioFeatures_genres.dropna(axis = 0, how = 'any', inplace = True)
audioFeatures_genres.reset_index(drop = True, inplace = True)
```

```python
rock_songs = audioFeatures_genres.loc[audioFeatures_genres['genres'] == 'rock'].copy()
rock_songs.reset_index(drop = True, inplace = True)
len(rock_songs)
```

```python
pop_songs = audioFeatures_genres.loc[audioFeatures_genres['genres'] == 'pop'].copy()
pop_songs.reset_index(drop = True, inplace = True)
len(pop_songs)
```

```python
cleaned_audioFeatures = audioFeatures_genres.drop('genres', axis = 1).copy()
cleaned_audioFeatures.head()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
scaler = MinMaxScaler()
scaled_audioFeatures = scaler.fit_transform(cleaned_audioFeatures)
scaled_audioFeatures
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split, cross_validate
```

```python
from nltk import ConfusionMatrix
from sklearn.metrics import classification_report
```

```python
k = 3
knn = KNeighborsClassifier(n_neighbors = k)
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_audioFeatures, audioFeatures_genres['genres'], 
                                                   shuffle = True, test_size = 0.3, random_state = 777)
```

```python
knn.fit(X_train, y_train)
predictions = knn.predict(X_test)
predictions
```

```python
confus_mat = ConfusionMatrix(list(y_test), list(predictions))
classifi_report = classification_report(y_test, predictions)

print(confus_mat)
print('\n\n')
print(classifi_report)
```

```python
from sklearn.metrics import precision_score, recall_score, make_scorer
custom_scorer = {'accuracy' : 'accuracy', 
                 'precision' : make_scorer(precision_score, average = 'micro'), 
                 'recall' : make_scorer(recall_score, average = 'micro')}
```

```python
scores = cross_validate(knn, scaled_audioFeatures, audioFeatures_genres['genres'], 
                        scoring = custom_scorer, cv = 10, return_train_score=False)
scores
```

```python
print(np.mean(scores['test_accuracy']))
print(np.mean(scores['test_precision']))
print(np.mean(scores['test_recall']))
```

### Big dataset of rock and pop songs (6435 rock songs & 1889 pop songs), NAs are replaced by mean of same genre

```python
audioFeatures = audioFeatureFrame(rock_pop_dataset.loc[:, 'audio_features'], audioFeatures_names)
audioFeatures.head()
```

```python
audioFeatures_genres = pd.concat([audioFeatures, rock_pop_dataset.loc[:,'genres']], axis = 1)
audioFeatures_genres.shape
```

```python
rock_means = audioFeatures_genres[audioFeatures_genres['genres']=='rock'].drop('genres', axis = 1).mean(axis = 0, skipna = True)
pop_means = audioFeatures_genres[audioFeatures_genres['genres']=='pop'].drop('genres', axis = 1).mean(axis = 0, skipna = True)
print(rock_means)
print(pop_means)
```

```python
for column in audioFeatures_genres.loc[:, :'mode_confidence'].columns.values:
    
    def fill_na(genres, val):
        if genres == 'rock':
            mean = rock_means[column]
        else:
            mean = pop_means[column]
            
        if np.isnan(val):
            return mean
        else:
            return val
        
    audioFeatures_genres[column] = audioFeatures_genres.apply(lambda row: fill_na(row['genres'], row[column]), axis =1)
```

```python
classes = ['rock', 'pop']
class_sizes = {genre : len(audioFeatures_genres[audioFeatures_genres['genres'] == genre]) for genre in classes}
print(class_sizes)
sizes = np.array(list(class_sizes.values()))
int(np.median(sizes))
```
