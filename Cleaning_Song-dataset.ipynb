# Pre-processing Songs dataset

```python
import pandas as pd
import numpy as np
```

```python
song_dataset = pd.read_json('MasterSongList.json')
print('Raw data size: ', song_dataset.shape)
song_dataset.head()
```

```python
def consolidateGenres (genre):
    if (len(genre) > 0):
        gen = genre[0]
        return gen.split(':')[0]
    else:
        return np.nan
    
song_dataset.loc[:, 'genres']= song_dataset['genres'].apply(consolidateGenres)
```

```python
song_dataset.iloc[23, :]
```

```python
song_dataset.loc[:, 'moods'] = song_dataset['moods'].apply(lambda x: ','.join(x))
song_dataset['moods']
```

```python
song_dataset[song_dataset['moods'].isnull()]
```

```python
audioFeatures_names = ['key', 'energy', 'liveliness', 'tempo', 'speechiness', 'acousticness', 'instrumentalness', 
                  'time_signature', 'duration', 'loudness', 'valence', 'danceability', 'mode', 
                  'time_signature_confidence', 'tempo_confidence', 'key_confidence', 'mode_confidence']
```

```python
def audioFeatureFrame (features_val, features_names):
    values = [val for val in features_val]
    frame = pd.DataFrame(values, columns = features_names)
    return frame
    
```

```python
audio_features = audioFeatureFrame(song_dataset.loc[:, 'audio_features'], audioFeatures_names)
audio_features
```

```python
mood_features = song_dataset.loc[:, 'moods'].str.get_dummies(sep=',')
mood_features
```

-------------------
## Assignment 1

```python
assignment_1_dataset = pd.concat([audio_features, mood_features, song_dataset.loc[:,'genres']], axis = 1)
assignment_1_dataset.head()
```

```python
assignment_1_dataset.to_csv("week1/assignment-1_dataset.csv", header = True, index = False)
```

--------------------
## Assignment 2

```python
genres_list = ['rock', 'jazz', 'dance', 'rap']
assignment_2_dataset = pd.DataFrame()

for genre in genres_list:
    
    row_filter = ((song_dataset['genres'] == genre) & (song_dataset['yt_views'] > 1000))
    
    audio_subset = audio_features.loc[row_filter, :]
    mood_genre_subset = song_dataset.loc[row_filter, ['moods', 'genres']]
    
    combined_subset = pd.concat([audio_subset, mood_genre_subset], axis = 1)
    combined_subset.dropna(axis = 0, how = 'any',inplace = True)
    
    balanced_subset = combined_subset.sample(n = 1500, axis = 0, random_state = 101)
    
    assignment_2_dataset = pd.concat([assignment_2_dataset, balanced_subset], axis = 0)
    
```

```python
print(assignment_2_dataset.shape)

assignment_2_dataset.head()
```

```python
assignment_2_dataset = assignment_2_dataset.sample(frac = 1, axis = 0, random_state = 101) #shuffling the data
assignment_2_dataset.reset_index(drop = True, inplace = True)
assignment_2_dataset
```

```python
assignment_2_dataset.to_csv('week2/assignment-2_dataset.csv', header = True, index = False)
```
