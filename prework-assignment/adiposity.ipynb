## 1. Set up environment and import data

```python
# for reading and working with data frame
import pandas as pd

# plotting modules
import seaborn as sns
import matplotlib.pyplot as plt

# import the data
data = pd.read_excel('BodyFat.xls')
print(data.head())
print("Data size: ", data.shape)
```

## 2. Have a look at the sample distribution of each variable (except record ID)

```python
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(15,15))

for ax, column in zip(axes.flat, data.columns[1:]):
    sns.distplot(data[column], ax=ax)
    
plt.show()
```

All the distributions are fairly normal. Pearson correlation method can be applied.
## 3. Obtain matrices of correlation coefficients

```python
corr_input = data.loc[:,'BODYFAT':] # Discard column of record ID 
corr = corr_input.corr()
corr_adiposity = corr.loc[['ADIPOSITY']]
```

## ... and visualize them as heatmaps

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10,10))
sns.heatmap(corr, cmap="Spectral", center = 0, ax=axes[0])
sns.heatmap(corr_adiposity, cmap="Spectral", center = 0, ax=axes[1])
plt.show()
```

### Measurement at abdomen is the feature that most strongly correlated with adiposity

```python
# Sorting the correlation coefficients
corr_adiposity_sorted = corr_adiposity.sort_values(by='ADIPOSITY', axis=1)
corr_adiposity_sorted
```

```python
def magnify():
    return [dict(selector="th",
                 props=[("font-size", "7pt")]),
            dict(selector="td",
                 props=[('padding', "0em 0em")]),
            dict(selector="th:hover",
                 props=[("font-size", "12pt")]),
            dict(selector="tr:hover td:hover",
                 props=[('max-width', '200px'),
                        ('font-size', '12pt')])
]

corr_adiposity_sorted.style.background_gradient(cmap= "Spectral", axis=1)\
    .set_properties(**{'max-width': '80px', 'font-size': '10pt'})\
    .set_caption("Interactive heatmap on correlations between different body characteristics and adiposity")\
    .set_precision(2)\
    .set_table_styles(magnify())
```

## 4. Split data into two: one with records where age under 45, and the other with records where age at or over 45

```python
under45 = data[data['AGE'] < 45]
over45 = data[data['AGE'] >= 45]
# Check the equality in sample size between two groups
print('Data size of records with age < 45: ', under45.shape)
print('Data size of records with age >= 45: ', over45.shape)
```

## 5. Correlations between characteristics and adiposity for each group

```python
corr_adiposity_under45 = under45.loc[:,'BODYFAT':].corr().loc[['ADIPOSITY']]
corr_adiposity_over45 = over45.loc[:,'BODYFAT':].corr().loc[['ADIPOSITY']]

# Visualize two correlation matrices as heatmaps
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(20, 5))
sns.heatmap(corr_adiposity_under45, cmap="Spectral", center = 0, ax=axes[0])
sns.heatmap(corr_adiposity_over45, cmap="Spectral", center = 0, ax=axes[1])
plt.show()
```

## 6. Find and visualize the changes in correlation when men pass 45

```python
change = corr_adiposity_under45 - corr_adiposity_over45
change_sorted = change.sort_values(by='ADIPOSITY', axis=1)
change_sorted
```

### When the body passes 45, correlation with measurement at the knee changes the most and is less than that when the body is under 45.

```python
change_sorted.style.background_gradient(cmap= "Spectral", axis=1)\
    .set_properties(**{'max-width': '80px', 'font-size': '10pt'})\
    .set_caption("Interactive heatmap on changes in correlation")\
    .set_precision(2)\
    .set_table_styles(magnify())
```
